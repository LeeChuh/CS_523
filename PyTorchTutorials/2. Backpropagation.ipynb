# Backpropagation in PyTorch

```python
import numpy as np
import torch 
```

```python
# loss = (y - x * w) ** 2
# y = 2
# x = 1
# w = 1
# seek d_loss / d_w

x = torch.tensor(1.0, requires_grad=True)
w = torch.tensor(1.0, requires_grad=True)
y = torch.tensor(2)

y_h = x * w
s = (y - y_h)
loss = s ** 2

# backward pass
loss.backward()
w.grad

### update weights
### next forward and backward propagagtion
```
